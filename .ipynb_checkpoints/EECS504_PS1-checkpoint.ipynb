```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def compute_homography(src_points, dst_points):
    """
    
    
    :
    src_points:  (n, 2)
    dst_points:  (n, 2)
    
    :
    H: 3x3
    """
    n = src_points.shape[0]
    A = np.zeros((2*n, 8))
    b = np.zeros(2*n)
    
    for i in range(n):
        x, y = src_points[i]
        x_prime, y_prime = dst_points[i]
        
        # Ab
        A[2*i] = [x, y, 1, 0, 0, 0, -x*x_prime, -y*x_prime]
        A[2*i+1] = [0, 0, 0, x, y, 1, -x*y_prime, -y*y_prime]
        
        b[2*i] = x_prime
        b[2*i+1] = y_prime
    
    # : H_params = (A^T A)^(-1) A^T b
    H_params = np.linalg.lstsq(A, b, rcond=None)[0]
    
    # 
    H = np.array([
        [H_params[0], H_params[1], H_params[2]],
        [H_params[3], H_params[4], H_params[5]],
        [H_params[6], H_params[7], 1]
    ])
    
    return H

def transform_points(points, H):
    """
    
    
    :
    points:  (n, 2)
    H: 3x3
    
    :
    transformed_points:  (n, 2)
    """
    # 
    homogeneous_points = np.column_stack([points, np.ones(len(points))])
    
    # 
    transformed_homogeneous = H @ homogeneous_points.T
    transformed_homogeneous = transformed_homogeneous.T
    
    # 
    transformed_points = transformed_homogeneous[:, :2] / transformed_homogeneous[:, 2:3]
    
    return transformed_points

def draw_virtual_yellow_line():
    """
    
    """
    # 
    img1 = cv2.imread('img1.jpg')  # 
    img2 = cv2.imread('img2.jpg')  # 
    
    img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
    
    # 
    # 
    src_points = np.array([
        [100, 200],   # 1
        [300, 200],   # 2  
        [100, 400],   # 3
        [300, 400],   # 4
        [150, 300],   # 1
        [250, 300]    # 2
    ])
    
    dst_points = np.array([
        [120, 180],   # 1
        [280, 190],   # 2
        [110, 380],   # 3  
        [290, 390],   # 4
        [160, 280],   # 5
        [240, 290]    # 6
    ])
    
    # 
    H = compute_homography(src_points, dst_points)
    
    # 33
    yellow_line_src = np.array([
        [50, 250],   # 
        [350, 250]   # 
    ])
    
    # 
    yellow_line_dst = transform_points(yellow_line_src, H)
    
    # 
    img2_with_line = img2_rgb.copy()
    start_point = tuple(yellow_line_dst[0].astype(int))
    end_point = tuple(yellow_line_dst[1].astype(int))
    
    # BGR
    cv2.line(img2_with_line, start_point, end_point, (0, 255, 255), 3)
    
    # 
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 3, 1)
    plt.imshow(img1_rgb)
    plt.title('Reference Image')
    plt.axis('off')
    
    plt.subplot(1, 3, 2)  
    plt.imshow(img2_rgb)
    plt.title('(target image')
    plt.axis('off')
    
    plt.subplot(1, 3, 3)
    plt.imshow(img2_with_line)
    plt.title('target image with yellow line')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()
    
    return img2_with_line, H

# 
if __name__ == "__main__":
    result_image, homography_matrix = draw_virtual_yellow_line()
    print("H:")
    print(homography_matrix)
```

```python
import numpy as np
import cv2
from scipy import ndimage

def demosaic(bayer_image):
    """
    Bayer - 
    
    :
        bayer_image: Bayer
        
    :
        rgb_image: 
    """
    height, width = bayer_image.shape
    rgb_image = np.zeros((height, width, 3), dtype=np.float32)
    
    # RGGB Bayer
    # : R G R G ...
    # : G B G B ...
    
    # 1: 
    for i in range(height):
        for j in range(width):
            if i % 2 == 0:  # 
                if j % 2 == 0:  #  - 
                    rgb_image[i, j, 0] = bayer_image[i, j]  # R
                else:  #  - 
                    rgb_image[i, j, 1] = bayer_image[i, j]  # G
            else:  # 
                if j % 2 == 0:  #  - 
                    rgb_image[i, j, 1] = bayer_image[i, j]  # G
                else:  #  - 
                    rgb_image[i, j, 2] = bayer_image[i, j]  # B
    
    # 2: 
    # 
    for i in range(height):
        for j in range(width):
            if rgb_image[i, j, 0] == 0:  # 
                neighbors = []
                # 
                for di in [-1, 1]:
                    for dj in [-1, 1]:
                        ni, nj = i + di, j + dj
                        if 0 <= ni < height and 0 <= nj < width:
                            if rgb_image[ni, nj, 0] > 0:  # 
                                neighbors.append(rgb_image[ni, nj, 0])
                
                if neighbors:
                    rgb_image[i, j, 0] = np.mean(neighbors)
    
    # 
    for i in range(height):
        for j in range(width):
            if rgb_image[i, j, 2] == 0:  # 
                neighbors = []
                for di in [-1, 1]:
                    for dj in [-1, 1]:
                        ni, nj = i + di, j + dj
                        if 0 <= ni < height and 0 <= nj < width:
                            if rgb_image[ni, nj, 2] > 0:
                                neighbors.append(rgb_image[ni, nj, 2])
                
                if neighbors:
                    rgb_image[i, j, 2] = np.mean(neighbors)
    
    # 
    for i in range(height):
        for j in range(width):
            if rgb_image[i, j, 1] == 0:  # 
                neighbors = []
                # 
                for dj in [-1, 1]:
                    nj = j + dj
                    if 0 <= nj < width and rgb_image[i, nj, 1] > 0:
                        neighbors.append(rgb_image[i, nj, 1])
                # 
                for di in [-1, 1]:
                    ni = i + di
                    if 0 <= ni < height and rgb_image[ni, j, 1] > 0:
                        neighbors.append(rgb_image[ni, j, 1])
                
                if neighbors:
                    rgb_image[i, j, 1] = np.mean(neighbors)
    
    return np.clip(rgb_image, 0, 255).astype(np.uint8)

# 
def process_orion_image():
    """
    orion.jpg
    """
    # orion.jpg
    bayer_image = cv2.imread('orion.jpg', cv2.IMREAD_GRAYSCALE)
    
    if bayer_image is None:
        print(": orion.jpg")
        return None
    
    # 
    color_image = demosaic(bayer_image)
    
    # 
    cv2.imwrite('orion_demosaiced.jpg', color_image)
    
    return color_image

# 
result = process_orion_image()
if result is not None:
    print(":", result.shape)
```
